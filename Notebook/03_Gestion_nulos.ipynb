## Nulos a analizar
- Subcuenta: es el código que se le dá a las cuentas contable, cambiaría a categórica. Tien 5% nulos
- Cliente: Identificador del cliente, cambiaría a categórica. Tiene 1,72% de nulos

```python
import pandas as pd
```

```python
df_o = pd.read_excel('../data/data_limpios.xlsx')
df_o.sample(2)
```

```python
df = df_o.copy()
```

```python
df.isnull().mean()*100
```

```python
df_nulos = df.loc[:,df.isnull().sum()>0]
df_nulos
```

```python
df_nulos['cliente'] = df_nulos['cliente'].astype('category')
```

```python
df_nulos.info()
```

Solo hay nulos en dos columnas categóricas

```python
for col in df_nulos.select_dtypes(include=['O','category']).columns:
    print(col.upper())
    display(df[col].value_counts()/df.shape[0]*100)
    print('--------')
```

### Gestion columna cliente. Como es tan sólo un 1,72% de los datos los sustuiré por desconocidos

```python
df['cliente'] = df['cliente'].cat.add_categories(['desconocido'])
df['cliente'] = df['cliente'].fillna('desconocido')
```

### Gestión columna descripcion_subcuen. Como sólo tiene un 5% de nulos haré lo mismo que con clientes sustituire por sin clasificar

```python
df['descripción_subcuen'] = df['descripción_subcuen'].fillna('sin_descripción')
```
