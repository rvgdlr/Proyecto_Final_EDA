```python
statistic, p_value =stats.shapiro(df['resultado'])

if p_value > 0.05:
    print(f'Para la columna Resultado los datos SI siguen una distribució normal')
else:
    print(f'Para la columna Resultado los datos NO siguen una distribució normal')
```

Como NO tiene una distribución normal tenemos que usar metodos con mediana. Test Main Whidneyt

```python


# 1. Agrupamos los datos: Creamos una lista de arrays, uno por cada línea de venta
grupos = [df[df['linea_venta'] == valor]['resultado'] for valor in df['linea_venta'].unique()]

# 2. Aplicamos Kruskal-Wallis (para comparar todos los grupos a la vez)
# El asterisco * desempaqueta la lista de grupos como argumentos individuales
statistic, p_value = stats.kruskal(*grupos)

print(f"Estadístico: {statistic:.4f}, P-value: {p_value:.4f}")

# 3. Interpretación correcta
if p_value > 0.05:
    print("No hay evidencias suficientes para decir que las medianas son diferentes.")
else:
    print("Existen diferencias significativas en al menos una de las medianas de los grupos.")
```
